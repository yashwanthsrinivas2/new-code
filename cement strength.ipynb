<b>Splitting of data<b/>

```python
target=df["Strength"]
original_data = df
original_data.drop(columns=["Strength"],inplace=True)
```

Now let us look at our input data

```python
original_data.head()
```

We now normalize the data 

```python
original_data=(original_data - original_data.mean())/original_data.std()
original_data.head()
```

Let us now build a baseline model for the analysis

```python
def neural_model():
    model= Sequential()
    model.add(Dense(10,activation="relu",input_shape=(8,)))##hidden layer with 10 nodes
    model.add(Dense(1,activation="relu"))## output layer with 1 node
    model.compile(optimizer ="adam",loss="mean_squared_error") ##optimizing and loss calculation
    return model
```

We will repeat the building og neural network for 50 times now

```python
mse_list=[]
i=1
X= original_data
y= target
while(i<=50):
    model =neural_model()
    Xtrain,xtest,ytrain,ytest = train_test_split(X,y,test_size=0.3,random_state=16)
    model.fit(Xtrain,ytrain,epochs=50)
    yhat=model.predict(xtest)
    mse_list.append(mean_squared_error(ytest,yhat)) ## calculating the mean squared errors  
    i=i+1
    
print("training done")
```

Now all loss values are stored in <b> mse_list</b> and we can calculate mean and std deviation from inbuilt functions

```python
import statistics
print("Mean is  "+str(statistics.mean(mse_list) ))
print("Standard deviation is  "+str(statistics.pstdev(mse_list)) )
```

```python
mean_normalized_50epoch_1hiddenlayer=statistics.mean(mse_list)
std_normalized_50epoch_1hiddenlayer=statistics.pstdev(mse_list)
```
